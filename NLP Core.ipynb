```python
paragraph = """Distressing issues that we are facing collectively. I think at times we feel, or we're made to feel, that we champion different causes. But for me, I see commonality. I think, whether we're talking about gender inequality or racism or queer rights or indigenous rights or animal rights, we're talking about the fight against injustice. We're talking about the fight against the belief that one nation, one people, one race, one gender or one species has the right to dominate, control and use and exploit another with impunity. I think that we've become very disconnected from the natural world, and many of us, what we're guilty of is an egocentric world view — the belief that we're the center of the universe. We go into the natural world, and we plunder it for its resources. We feel entitled to artificially inseminate a cow, and when she gives birth, we steal her baby, even though her cries of anguish are unmistakable. 
Then we take her milk that's intended for her calf, and we put it in our coffee and our cereal."""
```

```python
sentences = nltk.sent_tokenize(paragraph)

lemmatizer = WordNetLemmatizer()
```

```python
# Lemmatization
for i in range(len(sentences)):
    words = nltk.word_tokenize(sentences[i])
    newwords = [lemmatizer.lemmatize(word) for word in words]
    sentences[i] = ' '.join(newwords)
```

## Building the BOW Model

```python
import nltk
import re
```

```python
paragraph
```

```python
# dataset = []
```

```python
dataset = nltk.sent_tokenize(paragraph)

for i in range(len(dataset)):
    dataset[i] = dataset[i].lower()
    dataset[i] = re.sub(r'\W',' ',dataset[i])
    dataset[i] = re.sub(r'\s+',' ',dataset[i])
    
dataset
```

```python
# CREATING THE HISTOGRAM
```

```python
word2count = {}

for data in dataset:
    words = nltk.word_tokenize(data)
    for word in words:
        if word not in word2count.keys():
            word2count[word] = 1
        else:
            word2count[word] +=1
```

```python
word2count    # gives ('Word' , Qty)
```

```python
dataset #old
```

```python
# Most frequent N-largest Words
```

```python
import heapq
```

```python
  freq_words = heapq.nlargest(100, word2count,key=word2count.get)
```

```python
freq_words   # 'we' is the most frequent word then 'the' and so on
```

```python
# Using most frequent words to create a whole bag of words
```

```python
X = []   #each word apears asa  vector of leanth..... if it appears it's 1 otherwise 0
```

```python
for data in dataset:
    vector = []
    for word in freq_words:
        if word in nltk.word_tokenize(data):
            vector.append(1)     #pass
        else:
            vector.append(0)
    X.append(vector)
            
```

## TF IDF Model

- Term Frequency
- Inverse document frequency
- TF-IDF: TF*IDF

TF = No of occurance of a word in document % No. of words in that document

IDF = loge (No of documents % No of documents creating word)

TFIDF(word) = TF * IDF
